# split the dataset into test set and training set using stratified sampeling

```python
housing['income_cat']  = np.ceil(housing['median_income']/1.5)
```

```python
housing['income_cat'] = housing['income_cat'].apply(lambda x : x if x<5 else 5)
```

```python
housing['income_cat'].value_counts()/housing['income_cat'].count()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2,random_state=1)
```

```python
for train_index, test_index in sss.split(housing, housing['income_cat']):
    housing_train = housing.iloc[train_index].copy()
    housing_test = housing.iloc[test_index].copy()
```

```python
housing_train['income_cat'].value_counts()/housing_train['income_cat'].count()
```

```python
housing_test.drop(labels=['income_cat'], axis=1, inplace=True)
housing_train.drop(labels=['income_cat'], axis=1, inplace=True)
```
